```python
import json
import random

# 🔹 파일 경로 설정
input_path = r"C:\Users\asia\Desktop\파이널프로젝트\02_필요데이터셋\147.텍스트 윤리검증 데이터\01.데이터\1.Training\원천데이터\talksets-train-1\talksets-train-1.txt"
output_path = r"C:\Users\asia\Desktop\파이널프로젝트\03_데이터전처리\toxic_dataset.jsonl"

# 🔹 결과 리스트
samples = []

# 🔹 파일 읽기
try_encodings = ["utf-8-sig", "cp949", "utf-8"]

for enc in try_encodings:
    try:
        with open(input_path, "r", encoding=enc) as f:
            text = f.read()
        print(f"✅ 파일을 성공적으로 열었습니다! (인코딩: {enc})")
        break
    except UnicodeDecodeError:
        print(f"❌ 인코딩 {enc} 실패")

# 🔹 {} 단위로 대화 덩어리 분리
chunks = text.split("{")
chunks = [chunk.split("}")[0].strip() for chunk in chunks if "}" in chunk]

# 🔹 각 대화 처리
for idx, chunk in enumerate(chunks):
    utterances = chunk.split("|")
    utterances = [utt.strip() for utt in utterances if utt.strip()]  # 공백 제거

    if len(utterances) < 3:
        continue  # 문장이 3개 미만이면 skip

    if len(utterances) == 3:
        target_index = 2
    else:
        utterances = utterances[:4]  # 4개 이상이면 앞에서 4개만 사용
        target_index = random.choice([2, 3])

    sample = {
        "dialogue_id": f"{idx+1:03d}",
        "utterances": utterances,
        "target_index": target_index,
        "label": "toxic"
    }
    samples.append(sample)

# 🔹 2500개만 저장
samples = samples[:2500]

# 🔹 JSONL로 저장
with open(output_path, "w", encoding="utf-8") as f:
    for sample in samples:
        json.dump(sample, f, ensure_ascii=False)
        f.write("\n")

print(f"✅ 총 {len(samples)}개의 toxic 샘플이 저장되었습니다 → {output_path}")
```
