```python
import json
import random

# ğŸ”¹ íŒŒì¼ ê²½ë¡œ ì„¤ì •
input_path = r"C:\Users\asia\Desktop\íŒŒì´ë„í”„ë¡œì íŠ¸\02_í•„ìš”ë°ì´í„°ì…‹\147.í…ìŠ¤íŠ¸ ìœ¤ë¦¬ê²€ì¦ ë°ì´í„°\01.ë°ì´í„°\1.Training\ì›ì²œë°ì´í„°\talksets-train-1\talksets-train-1.txt"
output_path = r"C:\Users\asia\Desktop\íŒŒì´ë„í”„ë¡œì íŠ¸\03_ë°ì´í„°ì „ì²˜ë¦¬\toxic_dataset.jsonl"

# ğŸ”¹ ê²°ê³¼ ë¦¬ìŠ¤íŠ¸
samples = []

# ğŸ”¹ íŒŒì¼ ì½ê¸°
try_encodings = ["utf-8-sig", "cp949", "utf-8"]

for enc in try_encodings:
    try:
        with open(input_path, "r", encoding=enc) as f:
            text = f.read()
        print(f"âœ… íŒŒì¼ì„ ì„±ê³µì ìœ¼ë¡œ ì—´ì—ˆìŠµë‹ˆë‹¤! (ì¸ì½”ë”©: {enc})")
        break
    except UnicodeDecodeError:
        print(f"âŒ ì¸ì½”ë”© {enc} ì‹¤íŒ¨")

# ğŸ”¹ {} ë‹¨ìœ„ë¡œ ëŒ€í™” ë©ì–´ë¦¬ ë¶„ë¦¬
chunks = text.split("{")
chunks = [chunk.split("}")[0].strip() for chunk in chunks if "}" in chunk]

# ğŸ”¹ ê° ëŒ€í™” ì²˜ë¦¬
for idx, chunk in enumerate(chunks):
    utterances = chunk.split("|")
    utterances = [utt.strip() for utt in utterances if utt.strip()]  # ê³µë°± ì œê±°

    if len(utterances) < 3:
        continue  # ë¬¸ì¥ì´ 3ê°œ ë¯¸ë§Œì´ë©´ skip

    if len(utterances) == 3:
        target_index = 2
    else:
        utterances = utterances[:4]  # 4ê°œ ì´ìƒì´ë©´ ì•ì—ì„œ 4ê°œë§Œ ì‚¬ìš©
        target_index = random.choice([2, 3])

    sample = {
        "dialogue_id": f"{idx+1:03d}",
        "utterances": utterances,
        "target_index": target_index,
        "label": "toxic"
    }
    samples.append(sample)

# ğŸ”¹ 2500ê°œë§Œ ì €ì¥
samples = samples[:2500]

# ğŸ”¹ JSONLë¡œ ì €ì¥
with open(output_path, "w", encoding="utf-8") as f:
    for sample in samples:
        json.dump(sample, f, ensure_ascii=False)
        f.write("\n")

print(f"âœ… ì´ {len(samples)}ê°œì˜ toxic ìƒ˜í”Œì´ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤ â†’ {output_path}")
```
